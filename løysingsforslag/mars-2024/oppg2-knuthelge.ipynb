```python
import csv
import matplotlib.pyplot as plt

with open("run_ww_2020_w-PROVE.csv", "r") as file:
    content = csv.reader(file, delimiter=",")

    fields = next(content)

    rows = []
    for row in content:
        rows.append(row)
```

Del 1

```python
group_18_34 = 0
group_35_54 = 0
group_55_plus = 0

for row in rows:
    if row[6] == "18 - 34":
        group_18_34 += 1
    elif row[6] == "35 - 54":
        group_35_54 += 1
    elif row[6] == "55 +":
        group_55_plus += 1

print(f'I alderskategorien 18 - 34 er det {group_18_34} utøvere.')
print(f'I alderskategorien 35 - 54 er det {group_35_54} utøvere.')
print(f'I alderskategorien 55 + er det {group_55_plus} utøvere.')
```

Del 2

```python
plt.bar(("18 - 34", "35 - 54", "55 +"), (group_18_34, group_35_54, group_55_plus))
plt.xlabel("Aldersgrupper")
plt.ylabel("Antall utøvere")
```

Del 3

```python
total_time_per_week = [0]

weeknum = 1

week = rows[0][1]

for row in rows:
    if row[1] != week:
        weeknum += 1
        week = row[1]
        total_time_per_week.append(0)
    total_time_per_week[weeknum-1] += float(row[4])

max_time = 0
max_week = 0
for count, time in enumerate(total_time_per_week):
    if time > max_time:
        max_time = time
        max_week = count + 1

print(f"Uken med mest aktivitet var uke {max_week}.")
```

```python
weeknums = []
for i in range (1, 53):
    weeknums.append(str(i))

plt.figure(figsize=(15, 5))
plt.bar(weeknums, total_time_per_week)
plt.xlabel("Ukenummer")
plt.ylabel("Tid brukt for alle utøvere i minutter")
```
